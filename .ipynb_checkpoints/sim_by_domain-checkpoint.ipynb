```python
domain_dir = RES_DIR + 'job_prof/by_domain/' + 'new/'

def saveSim(sims, folder, name):
    sim_file = folder + 'sims/{}.csv'.format(clean(name))
    sims.to_csv(sim_file, index=False)
    print('\tSaved sims to file: {}'.format(sim_file))

def domainsInRange(min_n_title, max_n_title):
    q = '{} <= n_title < {}'.format(min_n_title, max_n_title)
    domains = by_domain_agg.query(q)['domain']
    msg = '# domains with # titles in range [{}, {}): {}'.format(min_n_title, max_n_title, len(domains))
    print(msg)
    return domains

def domainWiseSims(domains, k=5):
    for dom in domains:
        print('Domain: {}'.format(dom))
        chosen_titles = titlesIn(domain, title_df)
        
        sims = simsAmong(chosen_titles, doc_topic_distr, df, verbose)
        if not sims.empty:
            sims = sims.sort_values('topic_sim', ascending=False)
            saveSim(sims, domain_dir, dom)
        
            fig = viz(sims); fname = domain_dir + '{}_sims.pdf'.format(clean(dom)) 
            plt.savefig(fname)
            findTopkSimByDomain(domain=dom, sims=sims, k=k)
```
